```python
import pandas as pd
import numpy as np
data = 'https://raw.githubusercontent.com/FernandoGurgel/propina-no-more/master/bases/export_transparencia_2018.csv'
dataset = pd.read_csv(data, sep=";")
print("Extração concluída")

# import json
# with open('dadosOrgaos2016.json', 'w', encoding='utf-8') as file:
#     teste.to_json(file, force_ascii=False)
# print("Arquivo de 2016 salvo com sucesso!")
```

```python
unidades = dataset['UG_Sigla']
teste = dataset.groupby(['UG_Sigla','UG_Nome'])

sigla = dataset['UG_Sigla']
nome = dataset['UG_Nome']

dados = [sigla, nome]

periodico = []
dadosFormatados = []
print("Iniciando organização de dados...")
for y in range(len(dados[0])):
    for x in range(len(dados)):
        periodico.append(dados[x][y])    
    dadosFormatados.append(periodico)
    periodico = []
print("Terminou")
```

```python
  
print("Eliminando repetidos")
val = []
for i in range(len(dadosFormatados)): 
    if i == 0:
        temp = dadosFormatados[i][0]
        val.append(temp)
    else:
        if temp != dadosFormatados[i][0]:
            temp = dadosFormatados[i][0]
            val.append(temp)   
print("Terminou")
# import json
# with open('orgaos.json', 'w', encoding='utf-8') as file:
#     teste.to_json(file, force_ascii=False)
# print("Arquivo de 2016 salvo com sucesso!")
```
