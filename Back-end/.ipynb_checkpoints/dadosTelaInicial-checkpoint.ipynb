```python
import pandas as pd
import numpy as np
data = 'https://raw.githubusercontent.com/FernandoGurgel/propina-no-more/master/bases/export_transparencia_2018.csv'
dataset = pd.read_csv(data, sep=";")
print("Extração concluída")
```

```python
dataset
```

```python
dataset['Preco_Maximo'] = dataset['Preco_Maximo'].apply(lambda x: 0.0 if(x=='-') else float(x))

valoresLicitacoesPorOrgao = dataset.groupby(['UG_Sigla'], sort=True)['Preco_Maximo']

valoresLicitacoesPorOrgao_Total = valoresLicitacoesPorOrgao.agg('sum')

grupos = valoresLicitacoesPorOrgao_Total.sort_values(ascending=False)

print(grupos)

# groupby(['UG_Sigla'])['Preco_Maximo'].apply(pd.to_numeric)
# pd.to_numeric(teste,errors='ignore')  
# # teste.agg(['sum'])
# licitacoesPorOrgao.agg('sum')
```
