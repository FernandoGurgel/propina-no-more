```python
import pandas as pd
import numpy as np
data = 'https://raw.githubusercontent.com/FernandoGurgel/propina-no-more/master/bases/export_transparencia_2017.csv'
dataset = pd.read_csv(data, sep=";")
print("Extração concluída")
dataset
```

Ordenamento e tratamento dos valores numéricos

```python
#Ordenação dos valores numéricos e cálculo dos valores máximos
# dataset[(dataset['UG_Sigla'] == 'AADES') & (dataset['Situacao_Objeto'] == 'Aguardando Homologação') ]['Preco_Maximo'].values.astype(np.float32).sum()

dataset['Preco_Maximo'] = dataset['Preco_Maximo'].apply(lambda x: 0.0 if(x=='-') else float(x))
dataset['Preco_Maximo'].values.astype(np.float32).sum()
print("Dados numéricos convertidos em float")
```

```python
#Organização dos dados 
teste = dataset.groupby(['UG_Sigla','Situacao_Objeto']).size().to_frame('size')
print(teste)

#teste.agg('sum')
#print("Criação de um dataset")
```

```python
dataset['Preco_Maximo'] = dataset['Preco_Maximo'].apply(lambda x: 0.0 if(x=='-') else float(x))

compras = dataset.groupby(['UG_Sigla', 'Edital/Registro', 'Objeto', 'Empresa', 'Situacao_Objeto','Preco_Maximo'])['Preco_Maximo']

compras = compras.agg('sum')

grupos = compras.sort_values()

# print(compras)

t = dict(grupos)

# #print(t)

dados = []

for i in t.keys():
    dados.append({'ano':'2018','sigla':i[0], 'edital':i[1], 'objeto':i[2], 'empresa': i[3], 'situacao':i[4], 'valor':i[5]})

# # for i in t.keys():
# #     dados.append({'UG_SIGLA': i, 'VALOR_TOTAL_COMPRAS': t[i]}) 
    
#print(dados)

import json 
with open('valor_edital_2018.json', 'w', encoding='utf-8') as file:
    json.dump(dados, file,  ensure_ascii=False)
    
print("valor_edital_2018.json gerado")
```
